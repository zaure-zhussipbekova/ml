```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Score ~ C(Method)', data=data_vid).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
#Normality
import scipy.stats as stats

stats.shapiro(model.resid)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize= (10, 10))
ax = fig.add_subplot(111)data_vid

normality_plot, stat = stats.probplot(model.resid, plot= plt, rvalue= True)
ax.set_title("Probability plot of model residual's", fontsize= 20)
ax.set

plt.show()
```

```python
stats.levene(data_vid['Score'][data_vid['Method'] == 1],
             data_vid['Score'][data_vid['Method'] == 2])
```

```python
fig = plt.figure(figsize= (10, 10))
ax = fig.add_subplot(111)

ax.set_title("Box Plot of Score by Method", fontsize= 20)
ax.set

data = [data_vid['Score'][data_vid['Method'] == 1],
             data_vid['Score'][data_vid['Method'] == 2]]

ax.boxplot(data,
           labels= ['1', '2'],
           showmeans= True)

plt.xlabel("Method")
plt.ylabel("Score")

plt.show()
```
